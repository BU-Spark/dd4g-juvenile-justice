```python
ORIGINAL_DATASET_FILE = "../dataset/DESE School Discipline Data/Discipline 2018 all schools and districts - race - gender - spop.xlsx"
PROCESSED_DATASET_FILE = "../dataset/Discipline-2018.csv"
IS_2022 = False

if IS_2022:
    ECO_ROW = "Low Income"
else:
    ECO_ROW = "Economically Disadvantaged"
```

```python
import pandas as pd

df = pd.read_excel(ORIGINAL_DATASET_FILE)
```

```python
df["Gender"] = df["Gender"].fillna("Other")

df = df.dropna(subset=["Total Disciplined", "Total Eligible"], how="all")


def category_logic(row):
    if (
        row["Students w/ Disabilities"] != "With Disability"
        and row["English Learners"] != "English Learner"
        and row[ECO_ROW] != ECO_ROW
    ):
        return 0
    elif (
        row["Students w/ Disabilities"] == "With Disability"
        and row["English Learners"] != "English Learner"
        and row[ECO_ROW] != ECO_ROW
    ):
        return 1
    elif (
        row["Students w/ Disabilities"] != "With Disability"
        and row["English Learners"] == "English Learner"
        and row[ECO_ROW] != ECO_ROW
    ):
        return 2
    elif (
        row["Students w/ Disabilities"] != "With Disability"
        and row["English Learners"] != "English Learner"
        and row[ECO_ROW] == ECO_ROW
    ):
        return 3
    elif (
        row["Students w/ Disabilities"] == "With Disability"
        and row["English Learners"] == "English Learner"
        and row[ECO_ROW] != ECO_ROW
    ):
        return 4
    elif (
        row["Students w/ Disabilities"] != "With Disability"
        and row["English Learners"] == "English Learner"
        and row[ECO_ROW] == ECO_ROW
    ):
        return 5
    elif (
        row["Students w/ Disabilities"] == "With Disability"
        and row["English Learners"] != "English Learner"
        and row[ECO_ROW] == ECO_ROW
    ):
        return 6
    elif (
        row["Students w/ Disabilities"] == "With Disability"
        and row["English Learners"] == "English Learner"
        and row[ECO_ROW] == ECO_ROW
    ):
        return 7
    # else:
    #     print("!!!")

# Apply the function to each row in the dataframe
df["Category"] = df.apply(category_logic, axis=1)

```
