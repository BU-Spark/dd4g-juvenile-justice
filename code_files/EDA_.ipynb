```python
race_counts = data['Race/Ethnicity'].value_counts()
race_percentages = race_counts / len(data) * 100

# Plotting a bar chart for race/ethnicity distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=race_counts.index, y=race_counts.values)
plt.title('Distribution of Students by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Demographic analysis based on gender
gender_counts = data['Gender'].value_counts()
gender_percentages = gender_counts / len(data) * 100

# Plotting a pie chart for gender distribution
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Students by Gender')
plt.show()
```

```python
null_values = data[data['Total Disciplined'].isnull()]
```

```python
null_values.distinct('District Code')
```

```python
data.dropna(subset=['Total Disciplined'], inplace=True)
```

```python
data['Discipline Rate'] = data['Total Disciplined'] / data['Total Eligible']
```

```python
data.head()
```

```python
data.head(200)
```

```python
data['Category'] = data[['Students w/ Disabilities', 'English Learners', 'Economically Disadvantaged']].apply(
    lambda row: [cat for cat in row if pd.notnull(cat)], axis=1
)
```

```python
data.head(20)
```

```python
# Group by 'District Code' and 'Gender' and calculate the percentage of disciplined students for each category
grouped_data = data.groupby(['District Code', 'Gender'])['Total Disciplined', 'Total Eligible'].sum()
grouped_data['Discipline Rate'] = grouped_data['Total Disciplined'] / grouped_data['Total Eligible'] * 100

# Plot the percentage of disciplined students for each category
grouped_data['Discipline Rate'].plot(kind='bar', figsize=(10, 6))
plt.xlabel('District Code, Gender')
plt.ylabel('Discipline Rate (%)')
plt.title('Percentage of Disciplined Students by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the total disciplined students by district
district_disciplined = data.groupby('District Code')['Total Disciplined'].sum()

# Select the top N districts with the highest number of disciplined students
top_n_districts = district_disciplined.nlargest(10)

# Filter the data for the top N districts
filtered_data = data[data['District Code'].isin(top_n_districts.index)]

# Group by 'District Code' and 'Gender' and calculate the percentage of disciplined students for each category
grouped_data = filtered_data.groupby(['District Name', 'Gender'])['Total Disciplined', 'Total Eligible'].sum()
grouped_data['Discipline Rate'] = grouped_data['Total Disciplined'] / grouped_data['Total Eligible'] * 100

# Plot the percentage of disciplined students for each category
ax = grouped_data['Discipline Rate'].plot(kind='bar', figsize=(10, 6))
# ax.set_xticklabels(grouped_data.index, rotation=45)
plt.xlabel('District Name, Gender')
plt.ylabel('Discipline Rate (%)')
plt.title('Percentage of Disciplined Students by Category (Top {} Districts)'.format(10))

# Adjust the x-axis ticks
x_tick_labels = [f"{code}, {gender}" for code, gender in grouped_data.index]
plt.xticks(range(len(grouped_data)), x_tick_labels, rotation=45)


plt.show()
```
