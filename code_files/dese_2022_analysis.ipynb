```python
import pandas as pd
```

```python
discipline_data_2021 = pd.read_excel('../data/DESESchoolDisciplineData/Discipline2022allschoolsanddistricts-race-gender-spop.xlsx')
```

```python
discipline_data_2021.columns
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Gender', y='Total Disciplined', data=discipline_data_2021)
plt.title('Discipline Incidents by Gender')
plt.show()
```

```python
sns.boxplot(x='Race/Ethnicity', y='Total Disciplined', data=discipline_data_2021)
plt.figure(figsize=(30,10))
plt.title('Discipline Incidents by Race/Ethnicity')
plt.show()
```
